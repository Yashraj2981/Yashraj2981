_There are 16 ounces in a pound. Isn't the Imperial system great? Write a program that displays the equivalent grams for every integer value of ounces between 0 and 16 inclusive. Use a `for` loop._

```python
OUNCE_TO_GRAM = 28.3495

for ounces in range(0, 17):  # 0 to 16 inclusive
    grams = ounces * OUNCE_TO_GRAM
    print(f"{ounces} oz = {grams:.2f} g")
```

_This is a determinate loop, so a `for` was the right choice. But for comparison, write a version using a `while` loop below._

```python
OUNCE_TO_GRAM = 28.3495

ounces = 0
while ounces <= 16:
    grams = ounces * OUNCE_TO_GRAM
    print(f"{ounces} oz = {grams:.2f} g")
    ounces += 1
```

_Time to draw some shapes. Write a program that prompts the user to enter an integer, and then draws a square made up of `#` characters. The number of characters on each side should be the value the user entered._

```python
try:
    size = int(input("Enter the size of the square: "))
except:
    size = 5   # default value

if size < 2:
    print("#")
else:
    for i in range(size):
        if i == 0 or i == size - 1:
            print("#" * size)
        else:
            print("#" + " " * (size - 2) + "#")
```

_Finally, try a triangle. The value to be entered should be the height. The first row will have one `#`, the second two, and so on until the height._

```python
def draw_triangle():
    print("--- Triangle Program ---")

    # Get the height from the user
    try:
        height = int(input("Enter the height of the triangle: "))
    except ValueError:
        print("Invalid input. Please enter a whole number.")
        return

    # Basic validation
    if height <= 0:
        print("Height must be positive.")
        return

    # Draw the triangle
    for row in range(1, height + 1):
        # In row 1, print 1 '#'. In row 2, print 2 '#', etc.
        print("#" * row)
```

_Last one. Take one of your "shapes" programs, and extend so that it keeps running. After displaying a shape it should
prompt your user to enter `Y` if they want to see another. Carry on until they don't enter `Y`._

_Hint: This is easiest if you also use "Selection", but you can get by without. Up to you._

```python
def shapes_program_with_loop():
    print("--- Interactive Shapes Program ---")

    # Initialize the control variable for the loop
    keep_running = 'y'

    # Main loop: continues as long as the user enters 'y'
    while keep_running.lower() == 'y':
        print("\nWhat shape would you like to draw?")
        print("1: Rectangle")
        print("2: Triangle")

        choice = input("Enter 1 or 2: ")

        if choice == '1':
            # Run the rectangle drawing logic
            print("\n** Drawing a Rectangle **")
            try:
                width = int(input("Enter width: "))
                height = int(input("Enter height: "))
                if width > 0 and height > 0:
                    for i in range(height):
                        print("#" * width)
                else:
                    print("Dimensions must be positive.")
            except ValueError:
                print("Invalid input for dimensions.")

        elif choice == '2':
            # Run the triangle drawing logic
            print("\n** Drawing a Triangle **")
            try:
                height = int(input("Enter height: "))
                if height > 0:
                    for row in range(1, height + 1):
                        print("#" * row)
                else:
                    print("Height must be positive.")
            except ValueError:
                print("Invalid input for height.")

        else:
            print("Invalid shape choice. Please enter 1 or 2.")

        # Prompt the user to continue or exit
        keep_running = input("\nDo you want to see another shape? Enter 'y' to continue, or any other key to exit: ")

    print("\nProgram finished. Goodbye!")
```
